# Diccionarios
Un diccionario es una agrupacin de elementos que tienen una clave (key) y un valor asociado. Se puede utilizar la clave para recuperar el valor correspondiente del diccionario.

Los diccionarios se utilizan para guardar muchas partes de informacin en una sola variable. Los diccionarios se crean encerrando los pares clave: valor entre corchetes {}.

Los diccionarios son mutables; podemos cambiar, aadir o eliminar elementos una vez creado el diccionario.



```python
 bien_inmueble1 = {
  "constructora": "Cubic 33",
   "localizacin": "CDMX",
  "modelo": "PH5",
  "ao construccin": 2018,
  "ao entrega": 2021,
  "inteligente": True,
  "inmobiliaria": "Century 21",   
  "precio": 9567890
}
```

```python
type(bien_inmueble1)
```

Podemos crear diccionarios utilizando la funcin **dict**

```python
bien_inmueble2 = dict(constructura = "Cubic 33", localizacin= "CDMX", modelo =  "201", inteligente= False, precio= 4567980)
bien_inmueble2
```

Se puede acceder a las claves y sus valores correspondientes a travs de la notacin [].


```python
bien_inmueble1["modelo"]
```

```python
bien_inmueble2["inteligente"]
```

```python
#KeyError
bien_inmueble2["colonia"]
```

Para evitar esto, podemos utilizar el mtodo keys(). El mtodo keys() devolver una lista de todas las claves del diccionario.

```python
claves= bien_inmueble1.keys()
claves
```

Puedes comprobar si una clave est presente en un diccionario utilizando el operador **in**.

```python
"precio" in bien_inmueble2
```

```python
"direccion" in bien_inmueble1
```

El metodo values() nos permite revisar los valores asociados a las claves.

```python
valores = bien_inmueble1.values()
valores
```

El mtodo items() nos permite revisar los pares clave:valor. 

```python
pares = bien_inmueble1.items()
pares
```

Se puede cambiar el valor de un elemento haciendo referencia a su clave.

```python
bien_inmueble1["ao entrega"] = 2022
bien_inmueble1
```

El mtodo update() actualizar el diccionario con los elementos del argumento dado.
El argumento debe ser un diccionario, o un objeto con pares clave:valor.

```python
#Argumento diccionario
bien_inmueble2.update({"ao entrega": 2023, 
                      "inmobiliaria": "Navel"})
bien_inmueble2
```

```python
#Argumento objeto clave:valor
bien_inmueble2.update({"precio": 4596560})
```

Si queremos remover una clave y su calor correspondiente del diccionario, podemos utilizar el mtodo pop().

```python
bien_inmueble1.pop("constructora")
bien_inmueble1
```
