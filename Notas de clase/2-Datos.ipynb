# Métodos con strings
Existen diferentes funciones que permiten modificar o reordenar caracteres en Python. Revisaremos las más comunes. La lista completa de métodos se encuentra aquí: https://docs.python.org/3/library/stdtypes.html#string-methods

Si tenemos espacios en blanco sobrantes, podemos removerlos con la función **.strip( )**.

```python
#con espacios extra
string4 = "      Hamburguesa con queso       "
print(string4)
#quitamos los espacios sobrantes con .strip()
string4.strip()
```

También podemos buscar el índice (posición) de un caracter con la función **find( )**. 

```python
#Función find
string5 = "He vuelto de mi viaje"
sub_s1 = "vuelto"

#usando find para encontrar la posicion de una subcadena
buscador= string5.find(sub_s1)
print(buscador)
```

Y, de igual manera, podemos reemplazar un caracter con otro (como la función reemplazar de Word), con el método  **.replace( )**.

```python
#declaramos la variable string
ID = "TCGA56A12*"
#declaramos segunda variable string, que es a lo que deseamos reemplazar
sub_s2 = "8"
print(ID)
#reemplazamos caracteres indeseados con caracteres deseados
ID2= ID.replace("*", sub_s2)
print(ID2)
```

La función **join()** combina con cada elemento de una lista de strings formando una sola string.

```python
#una lista
palabras = ['Zophia', 'Data', 'Academy', 'es', 'genial']

#la unimos en una sola oración con join.
frase = " ".join(palabras)
print(palabras , "->", frase)
```

# NOTA
**None no es lo mismo que NaN**. NaN y None son tipos de datos diferentes en Python. NaN es la representación de No es un número (NaN), que es de tipo float numérico. None, como ya vimos antes,  es de tipo NoneType y es un objeto.
Sin embargo, cuando se trata de la detección y eliminación de valores perdidos, la libreria Pandas trata NaN y None de forma similar.

# Listas
Se define como un conjunto de valores ordenados. Pueden contener diferentes clases de variables y añadir, remover o cambiar los datos dentro de ellas.
Las listas son del tipo list. Sus elementos están encerrados entre corchetes [] y sus elementos son separados por comas ( , ) .

```python
plantas = ["hortensia", "helecho", "siempreviva","cerezo"]
```

```python
type(plantas)
```

Podemos hacer una lista con diferentes tipos de datos y variables, incluyendo a otra lista:

```python
lista = [6.89340,  5 <= 2 , variable_5, 78345, plantas]
lista
```

Podemos acceder a cierto(s) elemento(s) de la lista con la notación índice []. 

```python
plantas[0]
```

```python
plantas[2]
```

Si usas un índice que no está en la lista, Python arroja un IndexError.

```python
lista[6]
```

Puedes usar índices negativos, esto funcinará como si leyeras la lista de atras hacía delante.

```python
# acceder al elemento final de la lista.
plantas[-1]
```

Podemos acceder a múltiples elementos de la lista utilizando el **slicing de listas**. EStos métodos nos permiten obtener un fragmento (slice) de la estructura de datos. Para ello, utilizamos la notación **list[n : m]**, donde n es el índice de inicio del slice y m es el índice final (no inclusivo, es decir, no incluye el último elemento indicado). Así, podemos indicar qué elementos de la lista queremos obtener.

```python
# Elementos del al 3, porque no es inclusivo para el segundo elemento.
lista [1:4]
```

Si omitimos el valor de n, el slice iniciará al principio de la lista.

```python
lista [:4]
```

Si no agregamos el valor de m, el slice finalizará en el último elemento de la lista.

```python
lista [2:]
```

Que pasará si omitimos tanto n como m?

```python
lista [:]
```

Podemos utilizar el slicing para cambiar a los elementos de una lista. Si hacemos esto, los valores anteriores de la lista se pierden, ya que estamos actualizando la lista. Ejemplo: 

```python
lista_2 = [9357, "Data", 725.9, True, "Academy", 308123, None]
print(lista_2)
```

```python
#el slicing cambia la lista con los nuevos elementos indicados
lista_2= lista_2[0:2] 
print(lista_2)
```

Podemos utilizar la notación **len( )** para revisar el número de elementos que contiene una lista.

```python
len(lista)
```

```python
# Numero de elementos en la lista mas claro:
print("Numero de elementos en la lista:", len(lista))
```

En la línea de código anterior, utilizamos la función print() combinada con otro argumento, que es la función **len**. Después, le indicamos a Python de qué lista queremos que nos dé los elementos. 

Recordemos que las listas son estructuras de datos mutables. Por lo tanto, es posible cambiar los valores dentro de ellas, o el orden en el que se encuentran.
Se puede cambiar el valor de algún elemento de la lista con el operador de asignación (=) y la notación de índice [].

```python
plantas[0] = "pino"
print(plantas)
```

También es posible añadir otro elemento al final de la lista con **.append()**.

```python
plantas.append("girasol")
plantas
```

También se puede insertar un nuevo valor en un índice específico utilizando el método **.insert()**.


```python
plantas.insert(5, "orquidea")
plantas
```

Podemos añadir los elementos de la lista 1 a la lista 2 con el método **.extend()**.

```python
#construimos las listas
numeros = ["cincuenta", "cincuenta y uno"]
mas_numeros = ["cincuenta y dos ", "cincuenta y tres"]

# Extender la lista 
numeros.extend(mas_numeros)
print(numeros)
```

De manera alternativa, podemos extender la lista con el operador **+**.

```python
opcion2 = numeros + plantas
opcion2
```

El operador * repite los elementos de una lista un número n de veces.

```python
nones = [1,3,5,7,9]

#repetimos la lista 5 veces
nones = 5 * nones
print(nones)
```

Se puede borrar un elemento de la lista con **.remove()**.

```python
plantas.remove("girasol")
plantas
```

Para eliminar un elemento con índice específico, utilizamos el método **.pop()**. Pop también devuelve el elemento eliminado.
Si no especificamos un indice, pop eliminará el último elemento de la lista.

```python
plantas.pop(2)
```

```python
plantas
```

Para combinar dos o más listas, utilizamos el operador +. Esta operación también se denomina concatenación.

```python
plantas_2 = ["cícada", "musgo", "bugambilia"]
```

```python
mas_plantas = plantas + plantas_2
```

```python
mas_plantas
```

Uno de los métodos de listas más útiles es **.copy()**. Como su nombre lo indica, nos permite realizar una copia de una lista. De esta manera, mantenemos la lista original intacta. 

```python
lista_3 = [34, 'bugambilia', False, 4826, None]
print(lista_3)
```

```python
# copia los elementos de la lista a una nueva lista
lista_3_copia = lista_3.copy()
print(lista_3_copia)
```

```python
# modificamos la copia
lista_3_copia[-1] = 0
lista_3_copia
```

```python
#la lista original queda intacta
lista_3
```

Los tuples se encierran entre paréntesis ( ) y  sus elementos se separan con comas.

```python
mascotas = ("perro", "gato", "periquito")
```

```python
type(mascotas)
```

```python
len(mascotas)
```

```python
mascotas[0]
```

```python
mascotas[1]= "tortuga"
```

```python
animo= tuple(['feliz', 'triste', 'neutral'])
```

Los tuples pueden visualizarse con los métodos **count** e **index**.Estas funciones son las únicas incorporadas para los tuples.
 El método count() devuelve el número de veces que un valor especificado aparece en la tuple.
 El método index() encuentra la posición de la primera aparición del valor especificado y emite una excepción si no se encuentra el valor.

```python
animo.count("feliz")
```

```python
animo.index("neutral")
```

# Diccionarios
Un diccionario es una agrupación de elementos que tienen una clave (key) y un valor asociado. Se puede utilizar la clave para recuperar el valor correspondiente del diccionario.

Los diccionarios se utilizan para guardar muchas partes de información en una sola variable. Los diccionarios se crean encerrando los pares clave: valor entre corchetes {}.

Los diccionarios son mutables; podemos cambiar, añadir o eliminar elementos una vez creado el diccionario.



```python
 bien_inmueble1 = {
  "constructora": "Cubic 33",
   "localización": "CDMX",
  "modelo": "PH5",
  "año construcción": 2018,
  "año entrega": 2021,
  "inteligente": True,
  "inmobiliaria": "Century 21",   
  "precio": 9567890
}
```

```python
type(bien_inmueble1)
```

Podemos crear diccionarios utilizando la función **dict**

```python
bien_inmueble2 = dict(constructura = "Cubic 33", localización= "CDMX", modelo =  "201", inteligente= False, precio= 4567980)
bien_inmueble2
```

Se puede acceder a las claves y sus valores correspondientes a través de la notación [].


```python
bien_inmueble1["modelo"]
```

```python
bien_inmueble2["inteligente"]
```

```python
#KeyError
bien_inmueble2["colonia"]
```

Para evitar esto, podemos utilizar el método keys(). El método keys() devolverá una lista de todas las claves del diccionario.

```python
claves= bien_inmueble1.keys()
claves
```

Puedes comprobar si una clave está presente en un diccionario utilizando el operador **in**.

```python
"precio" in bien_inmueble2
```

```python
"direccion" in bien_inmueble1
```

El metodo values() nos permite revisar los valores asociados a las claves.

```python
valores = bien_inmueble1.values()
valores
```

El método items() nos permite revisar los pares clave:valor. 

```python
pares = bien_inmueble1.items()
pares
```

Se puede cambiar el valor de un elemento haciendo referencia a su clave.

```python
bien_inmueble1["año entrega"] = 2022
bien_inmueble1
```

El método update() actualizará el diccionario con los elementos del argumento dado.
El argumento debe ser un diccionario, o un objeto con pares clave:valor.

```python
#Argumento diccionario
bien_inmueble2.update({"año entrega": 2023, 
                      "inmobiliaria": "Navel"})
bien_inmueble2
```

```python
#Argumento objeto clave:valor
bien_inmueble2.update({"precio": 4596560})
```

Si queremos remover una clave y su calor correspondiente del diccionario, podemos utilizar el método pop().

```python
bien_inmueble1.pop("constructora")
bien_inmueble1
```

# Otros métodos 

El método get() devuelve el valor del elemento con la clave especificada.

```python
bien_inmueble2.get("modelo")
```


El método clear() vacía el diccionario.


```python
bien_inmueble2.clear()
```

La función copy() realiza una copia del diccionario. Al igual que en las listas, tener una copia de un diccionario nos permite modificarlo y además tener el original intacto.

```python
bien_inmueble1.copy()
```
