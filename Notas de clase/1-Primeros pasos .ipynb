## Podemos utilizar variables para resolver problemas numéricos

Como los valores se almacenan en las variables, podemos llamarlas directamente para resolver algunas operaciones. 

Por ejemplo:  
Una agencia de viajes quiere llevar un grupo de turistas a las Egipto. Para ello, primero debe calcular los gastos de operación. El vuelo de avión cuesta 5.000 dólares de ida y 5.500 dólares de vuelta. La estadía en Egipto, 1.000 dólares por semana por habitación. Hasta ahora, 4 personas (1 pareja y 2 personas que viajan solas) han reservado el viaje. La empresa revisa los precios de la competencia, y obervan que cobran 2500 dólares por persona para un viaje similar. Podemos resolver estas preguntas creando variables y utilizando las operaciones aritméticas adecuadas.

# Contrastando fechas con los comparadores 
Una de las aplicaciones más útiles de los comparadores de Python es la comprobación de fechas, para saber si algún evento ocurrió antes o después de manera rápida.  Las fechas en Python no son un tipo de dato propio, pero podemos importar un módulo llamado datetime para trabajar con fechas como objetos de fecha.
Aquí unas celdas de ejemplo.

````python
#Importamos el módulo datetime
import datetime
````

Veremos qué es un módulo de  Python más adelante. Por el momento, solo necesitas saber que un módulo es un archivo que contiene funciones útiles para nuestra aplicación. En este caso, la librería datetime proporciona clases para manipular fechas y horas. Lo necesitamos porque Python no tiene este tipo de variables de tiempo definidas de manera intínseca. 

````python
#No hay tipo de variable tiempo
tiempo = (2020, 5, 10)
type(tiempo)
````

````python
tiempo2 = 2020/6/6
type(tiempo2)
````

````python
#comparamos 
fecha1>fecha2
````

## Condiciones y operadores lógicos en Python


Los operadores lógicos **and**, **or** y **not** funcionan con condiciones y valores **True** y **False** (también llamados booleanos). **And** y **or** funcionan con dos estados, mientras que not funciona con un solo estado.

El operador **and** devuelve *True* cuando ambas condiciones se evalúan como *True*. En caso de no ser así, devuelve *False*.


````python
#Con and ambas deben ser verdaderas para obtener True
num8 and num4 <=100
````

````python
not num1 == 5
````

````python
#Con or, al menos una es verdadera para obtener True
(num5 or num2) != 100
````

````python
(num7 or num4) > 78 
````
