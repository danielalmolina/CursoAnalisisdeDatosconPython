## Funcin agg
Esta funcin nos permite aadir distintas funciones de agregacin (sum, max, min, counts, etc), utilizando una o ms operaciones sobre un eje especificado (filas o columnas). Existen mltiples maneras de llamar a una funcin de agregaci dentro de agg. Por ejemplo, podemos pasar una lista de funciones para aplicar a una o ms columnas de datos.

Qu pasa si queremos realizar el anlisis slo en un subconjunto de diferentes columnas? Hay otras dos opciones para las agregaciones: usar un diccionario o una agregacin con nombre. Revisaremos cada uno de estos casos. 
Documentacin: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html

```python
#pasando una lista para evaluar diferentes funciones de agregacin sobre una columna
covid_world['total_deaths'].agg(['sum', 'count'])
```

```python
#pasando un diccionario para evaluar un subconjunto de mltiples columnas
covid_world.agg({'new_cases' : ['sum', 'min', 'max'], 'new_deaths' : ['min', 'max', 'sum']})
```

## Funcin groupby
La funcin **groupby** de Pandas permite dividir un DataFrame en grupos, aplicar una funcin a cada categora de forma independiente, y luego combinar los resultados de nuevo. Esto se denomina el patrn "dividir-aplicar-combinar", y es una poderosa herramienta para analizar los datos a travs de diferentes categoras. Ya hemos hecho esto anteriormente! Cuando revisamos los datos de la India, utilizamos 
**covid_world[covid_world.location == "India"]**. 
Como veremos a continuacin, **groupby** es ms eficiente, informativa y fcil de usar. 
Documentacin: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html


```python
#Filtrado del df por cada continente y por el numero de casos nuevos. Finalmente, se suman los valores de la columna nuevos casos
covid_world.groupby('continent').new_cases.sum()
```

```python
#Filtrado del df por cada continente y por el numero de casos nuevos. Finalmente, se toman los valores mximos de la columna nuevos casos
covid_world.groupby('continent').new_cases.max()
```

*Tip*: podemos parafrasear **groupby** por **cada uno**. Ej: anlisis de nmero de casos nuevos por cada uno de los continentes.

```python
#Combinando agg con groupby
#Pasando una lista para aplicar simultneamente varias funciones de agregacin
covid_world.groupby('continent').new_cases.agg(['count', 'min', 'max'])
```

# Merge 

Tpico problema de data analyst (o data scientist): queremos unir dos dataframes. No solo eso, sino que queremos que uno de esos dataframes haga "match" con el otro, quiz reemplazando alguna columna o conservndola Copiar y pegar en Excel? Tal vez. Pero, qu pasa si tienes miles o millones de registros? Para eso existen las funciones merge y join de Pandas. 

* pd.merge(df1, df2): permite realizar uniones mltiples en varias columnas.


Tomado de Medium

```python
#Primer dataframe
df1 = pd.DataFrame({'key': ['a', 'b', 'u1', 'a'], 'value': [1, 2, 3, 5]})
df1
```

```python
#Segundo dataframe
df2 = pd.DataFrame({'key': ['a', 'b', 'u2', 'a'], 'value': [5, 6, 7, 8]})
df2
```

Intentemos unir ambos dataframes. 

```python
pd.merge(df1,df2)
```

Merge busca valores idnticos en las columnas **key** y **value**. Por lo tanto, esta fusin nos da un dataframe casi vaco porque la mayora de los valores de las columnas no coinciden, y solo nos devuelve a 5, la nica fila que s coincide. 
*  Vamos a utilizar una columna particular para unir ambos dataframes. En este caso, utilizaremos la columna **key**.

```python
#inner es el valor por default de la funcin, pero queremos hacerlo explcito aqu.
df1.merge(df2, how="inner", on="key")
```

* x representa al primer elemento a unir (df1).
* y representa al segundo elemento a unir (df2).

```python

df1.merge(df2, how="outer", on="key")
```

```python
df1.merge(df2, how="left", on="key")
```

```python
df1.merge(df2, how="right", on="key")
```

```python
df3 = pd.DataFrame({'key': ['a', 'b', 'u2', 'a'], 'value': [5, 6, 7, 8], 'C': ['sol', 'luna', 'marte', 'venus']})
df3
```

```python
#podemos realizar la unin pasando una lista con  mltiples columnas
df2.merge(df3, on=['key', 'value'])
```
