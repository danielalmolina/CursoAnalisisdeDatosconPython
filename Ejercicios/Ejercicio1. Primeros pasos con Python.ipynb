Toma en cuenta que para calcular el rendimiento, debes utilizar el cociente = (Ganancias anuales netas / inversión) * 100.

Recuerda declarar variables en cada caso y realizar las operaciones adecuadas con esas variables, no lo hagas con los valores numéricos.

```python
costo_departamento = 3450000
alquiler = 12500
costos_mensuales = 3000
```

```python
ganancia = alquiler - costos_mensuales
ganancia
```

```python
ganancias_anuales_netas = ganancia * 12
ganancias_anuales_netas
```

```python
rendimiento_inversión = (ganancias_anuales_netas / costo_departamento) * 100
rendimiento_inversión
```
