```python
# Funcion 8. Crearemos un nuevo DataFrame para saber en que año nacieron la mayor
# parte de nuestros clientes.
gen_clientes = costumers.loc[:,['customer_id','year_of_birth ']]
gen_clientes
```

```python
# Funcion 9. Ahora usaremos una funcion que agrupe nuestro anterior DataFrame
# de tal forma que nos muestre el año de nacimiento de nuestros clientes.
gen_clientes.groupby('year_of_birth ').agg('count')
```

De acuerdo a lo anterior, encontramos que lla mayor cantidad de clientes nacieron en 1957 con 27 clientes (65 años), seguido de 1982 con 26 clientes (40 años), en tercer luegar se encuentra 1981 con 25 clientes (39 años) y en cuarto lugar 1980 (38 años) con 24 clientes. La mayoria de nuestros clientes son mayores de edad y problablemente poseen una vida adulta estable.

Para ello seguiremos investigando mas sobre ellos ...

```python
# Funcion 10. Haremos un slice diferente de nuestros clientes ahora con el nivel de educacion.
level_edu_clientes= costumers[['year_of_birth ','educational_level','annual_income']]
level_edu_clientes
```

```python
# Función 11. Ahora agrupamos por el 'nivel de educacion' y hacemos una agregacion por 'ingreso anual'.
level_edu_clientes.groupby('educational_level').annual_income.agg(['count', 'min', 'max'])
```

Encontramos que 249 de nuestros clientes estan graduados de la universidad y como minimo tienen un ingreso anual de 2,447 dolares y un ingreso anual maximo de 102,692 dolares. Seguido de 111 clientes con un PhD (Doctorado) donde el ingreso minimo anual es de 5,648 dolares y un ingreso maximo anual de 157,243 dolares.

Ahora buscaremos saber si tenemos mas ventas en linea o en la tienda...

```python
# Funcion 12. Haremos un tercer slice 
type_of_purchases = costumers[['year_of_birth ','online_purchases','store_purchases']]
type_of_purchases
```

```python
# Funcion 13. Volveremos a agrupar y agregar por el tipo de compras
type_of_purchases.groupby('year_of_birth ').agg({'online_purchases' : ['sum'], 'store_purchases' : ['sum']})
```

Nos podemos dar cuenta que de acuerdo a nuestros grueso de clientes que habiamos presentado anteriormente son lo que realizaron la mayor cantidad de compras, sin embargo los del año 1957 son los que hacen la mayoria de sus comprar en linea, a comparacion de los del año 1980 a 1983 los cuales realizan casi la misma cantidad de compras tanto en linea como en tienda.

```python
# Funcion 14. Sumar por columnas, en este caso sumar todas las compras online
total_online = type_of_purchases['online_purchases'].sum()
total_online
```

```python
# Funcion 15.Sumar por columnas, en este caso sumar todas las compras en tienda 
total_store = type_of_purchases['store_purchases'].sum()
total_store
```

Y tambien descubrimos que aun la mayoria de nuestras compras siguen siendo realizadas en la tienda con 2884 compras a comparacion de 1981 compras online. Y muy probablemente se deba a las generaciones de nuestros clientes.
